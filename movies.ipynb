```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import r2_score  ##r^2
import matplotlib.pyplot as plt

regr = linear_model.LinearRegression()

datos = pd.read_csv('C:\home\jupyter\data\IMDB-Movie-Data.csv')
df = pd.DataFrame(datos)
df

```

```python
x = df['Votes']
y= df['Rating']
```

```python
X = x[:,np.newaxis]  #da formato de arreglo en numpy a los datos
print(X)  
```

```python
regr.fit(X,y)
print(regr.coef_)
m=regr.coef_[0]  ##pendiente
b=regr.intercept_  ## intercepciÃ³n con el eje y
y_p = m*X+b  #predicted value

print('y={0}*x + {1}'.format(m,b))
```

```python
print(regr.predict(X[0:5]))
print('El valor de r^2: ', r2_score(y,y_p))  #comparaciÃ³n de valor real y y predicciÃ³n y_p
```
